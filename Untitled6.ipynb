```python
# -----------------------------
# 1. Imports
# -----------------------------
import tensorflow as tf
from tensorflow.keras.applications import ResNet50, EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Lambda, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing import image
import numpy as np
import random
import os
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.manifold import TSNEimport matplotlib.pyplot as plt
from seaborn import sns
# -----------------------------
# 2. Dataset setup
# -----------------------------
dataset_path = "animals_dataset/animals/animals"

# Get class folders and images
classes = os.listdir(dataset_path)
class_to_images = {cls: [os.path.join(dataset_path, cls, img) for img in os.listdir(os.path.join(dataset_path, cls))] for cls in classes}
all_labels = list(class_to_images.keys())

# -----------------------------
# 3. Image preprocessing
# -----------------------------
IMG_SIZE = 224

def preprocess_img(img_path, backbone="resnet50"):
    img = image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    if backbone.lower() == "resnet50":
        img = tf.keras.applications.resnet50.preprocess_input(img)
    elif backbone.lower() == "efficientnetb0":
        img = tf.keras.applications.efficientnet.preprocess_input(img)
    return img[0]

# -----------------------------
# 4. Data generator (on-the-fly pairs)
# -----------------------------
def siamese_data_generator(class_to_images, all_labels, batch_size=16, backbone="resnet50"):
    while True:
        x1_batch = np.zeros((batch_size*2, IMG_SIZE, IMG_SIZE, 3), dtype=np.float32)
        x2_batch = np.zeros((batch_size*2, IMG_SIZE, IMG_SIZE, 3), dtype=np.float32)
        y_batch = np.zeros((batch_size*2,), dtype=np.float32)

        for i in range(batch_size):
            # Positive pair
            label = random.choice(all_labels)
            img1, img2 = random.sample(class_to_images[label], 2)
            x1_batch[2*i] = preprocess_img(img1, backbone)
            x2_batch[2*i] = preprocess_img(img2, backbone)
            y_batch[2*i] = 1.0

            # Negative pair
            label1, label2 = random.sample(all_labels, 2)
            img1 = random.choice(class_to_images[label1])
            img2 = random.choice(class_to_images[label2])
            x1_batch[2*i+1] = preprocess_img(img1, backbone)
            x2_batch[2*i+1] = preprocess_img(img2, backbone)
            y_batch[2*i+1] = 0.0

        yield (x1_batch, x2_batch), y_batch

# -----------------------------
# 5. Siamese network model
# -----------------------------
def build_siamese_model(backbone="resnet50"):
    # Choose backbone
    if backbone.lower() == "resnet50":
        base_model = ResNet50(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
    elif backbone.lower() == "efficientnetb0":
        base_model = EfficientNetB0(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))

    base_model.trainable = False  # Freeze backbone

    # Feature extraction network
    inp = Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base_model(inp)
    x = GlobalAveragePooling2D()(x)
    x = Dense(256, activation="relu")(x)
    feature_model = Model(inp, x)

    # Siamese inputs
    input_a = Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    input_b = Input(shape=(IMG_SIZE, IMG_SIZE, 3))

    feat_a = feature_model(input_a)
    feat_b = feature_model(input_b)

    # L1 distance
    l1_distance = Lambda(lambda tensors: tf.abs(tensors[0] - tensors[1]))([feat_a, feat_b])
    out = Dense(1, activation="sigmoid")(l1_distance)

    model = Model(inputs=[input_a, input_b], outputs=out)
    model.compile(optimizer=Adam(1e-4), loss="binary_crossentropy", metrics=["accuracy"])
    return model , feature_model

# -----------------------------
# 6. Training function
# -----------------------------
def train_siamese(backbone_name="resnet50", epochs=10, batch_size=16, steps_per_epoch=100, validation_steps=20):
    print(f"\n--- Training Siamese Network with {backbone_name.upper()} ---\n")

    # Unpack both models
    model, feature_model = build_siamese_model(backbone=backbone_name)

    history = model.fit(
        siamese_data_generator(class_to_images, all_labels, batch_size=batch_size, backbone=backbone_name),
        steps_per_epoch=steps_per_epoch,
        validation_data=siamese_data_generator(class_to_images, all_labels, batch_size=batch_size, backbone=backbone_name),
        validation_steps=validation_steps,
        epochs=epochs
    )

    return model, feature_model, history

# -----------------------------
# 7. Train ResNet50
# -----------------------------
model_resnet, feature_resnet, history_resnet = train_siamese(backbone_name="resnet50", epochs=5, batch_size=16)

# -----------------------------
# 8. Train EfficientNetB0
# -----------------------------
hmodel_resnet, feature_resnet, istory_effnet = train_siamese(backbone_name="efficientnetb0", epochs=5, batch_size=16)

# -----------------------------
# 9. Compare results
# -----------------------------
plt.figure(figsize=(12,5))
plt.plot(history_resnet.history['accuracy'], label='ResNet50 Train Acc')
plt.plot(history_resnet.history['val_accuracy'], label='ResNet50 Val Acc')
plt.plot(history_effnet.history['accuracy'], label='EffNetB0 Train Acc')
plt.plot(history_effnet.history['val_accuracy'], label='EffNetB0 Val Acc')
plt.title('Training vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

plt.figure(figsize=(12,5))
plt.plot(history_resnet.history['loss'], label='ResNet50 Train Loss')
plt.plot(history_resnet.history['val_loss'], label='ResNet50 Val Loss')
plt.plot(history_effnet.history['loss'], label='EffNetB0 Train Loss')
plt.plot(history_effnet.history['val_loss'], label='EffNetB0 Val Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()


```
